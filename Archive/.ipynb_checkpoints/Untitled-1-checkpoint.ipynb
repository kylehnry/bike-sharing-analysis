```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import geopandas as gpd
from shapely.geometry import Point
```

```python
# Load data from CSV file
file_path = 'Cleaned_data_final.csv'
data = pd.read_csv(file_path)

# Display the first few rows to understand the structure
print(data.head())

```

Convert date columns to datetime format, filter trips that occurred between 09:00-10:00 and 18:00-19:00, and separate station data from trip data.

```python
# Convert date columns to datetime
data['Start date'] = pd.to_datetime(data['Start date'])
data['End date'] = pd.to_datetime(data['End date'])

# Extract hour from 'Start date' for filtering
data['start_hour'] = data['Start date'].dt.hour

# Filter trips to only include those that took place between 09:00-10:00 and 18:00-19:00
filtered_trips = data[(data['start_hour'] == 9) | (data['start_hour'] == 18)]

# Separate stations data (latitude and longitude) from trips data
stations_df = data[['Start station number', 'lat_start', 'lon_start']].drop_duplicates()
stations_df.columns = ['station_id', 'latitude', 'longitude']

# Trips data
trips_df = filtered_trips[['Bike number', 'Start station number', 'End station number', 'Total duration (ms)', 'start_hour']]
trips_df.columns = ['bike_number', 'start_station', 'end_station', 'trip_duration', 'start_hour']
```

Merge trip data with station locations to include geographic coordinates for both start and end stations, and calculate trip counts and average trip duration for each station.

```python
# Merge trip data with station locations for start and end stations
trips_with_locations = trips_df.merge(stations_df, left_on='start_station', right_on='station_id', suffixes=('', '_start'))
trips_with_locations = trips_with_locations.merge(stations_df, left_on='end_station', right_on='station_id', suffixes=('', '_end'))

# Calculate trip counts and average trip duration for each station
trip_stats = trips_with_locations.groupby('start_station').agg({
    'trip_duration': ['count', 'mean']
}).reset_index()
trip_stats.columns = ['station_id', 'trip_count', 'avg_trip_duration']

# Merge with station locations
station_stats = stations_df.merge(trip_stats, on='station_id', how='left').fillna(0)
```

Use the Elbow method to determine the optimal number of clusters and apply K-Means clustering to group stations based on their geographic coordinates.

```python
# Features for clustering
X = station_stats[['latitude', 'longitude']]

# Determine the optimal number of clusters using the Elbow method
sse = []
for k in range(1, 11):
    kmeans = KMeans(n_clusters=k, random_state=42).fit(X)
    sse.append(kmeans.inertia_)

plt.plot(range(1, 11), sse)
plt.xlabel('Number of clusters')
plt.ylabel('SSE')
plt.title('Elbow Method for Optimal k')
plt.show()

# Apply K-Means clustering
optimal_k = 4  # Example value, determine from Elbow method
kmeans = KMeans(n_clusters=optimal_k, random_state=42)
station_stats['cluster'] = kmeans.fit_predict(X)
```

Convert the clustered data into a GeoDataFrame and plot the clusters on a map to visualize the spatial distribution of bike sharing stations.

```python
# Convert to GeoDataFrame for visualization
geometry = [Point(xy) for xy in zip(station_stats['longitude'], station_stats['latitude'])]
gdf = gpd.GeoDataFrame(station_stats, geometry=geometry)

# Plot clusters
gdf.plot(column='cluster', cmap='viridis', legend=True)
plt.title('Bike Sharing Stations Clusters')
plt.show()
```

```python
# Save the results to a new Excel file
output_file_path = 'Clustered_Stations.xlsx'
with pd.ExcelWriter(output_file_path) as writer:
    station_stats.to_excel(writer, sheet_name='Station_Clusters', index=False)
    gdf.to_excel(writer, sheet_name='GeoData', index=False)
```

This cell calculates summary statistics for trip counts and average trip duration within each cluster, to understand the central tendencies and variability of each cluster.

```python
cluster_summary = station_stats.groupby('cluster').agg({
    'trip_count': ['mean', 'median', 'std', 'sum'],
    'avg_trip_duration': ['mean', 'median', 'std']
}).reset_index()

print(cluster_summary)
```

This cell creates boxplots to compare trip counts and average trip durations across clusters, identify differences and similarities between clusters.

```python
import seaborn as sns

# Boxplot of trip counts by cluster
plt.figure(figsize=(10, 6))
sns.boxplot(x='cluster', y='trip_count', data=station_stats)
plt.title('Trip Counts by Cluster')
plt.show()

# Boxplot of average trip duration by cluster
plt.figure(figsize=(10, 6))
sns.boxplot(x='cluster', y='avg_trip_duration', data=station_stats)
plt.title('Average Trip Duration by Cluster')
plt.show()
```

This cell identifies outlier stations within each cluster based on trip counts, helping you spot stations with unusual activity.

```python
# Find stations that are outliers in trip counts within each cluster
from scipy import stats
import numpy as np

# Calculate z-scores for trip_count within each cluster
station_stats['trip_count_zscore'] = station_stats.groupby('cluster')['trip_count'].transform(lambda x: stats.zscore(x))

# Identify outliers based on z-score threshold
outliers = station_stats[station_stats['trip_count_zscore'].abs() > 2]

# Display the outliers
from IPython.display import display, HTML

display(HTML(outliers.to_html()))
```
