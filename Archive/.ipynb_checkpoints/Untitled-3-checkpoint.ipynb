```python
import pandas as pd
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point

# Load data
file_path = 'Cleaned_data_final.csv'
data = pd.read_csv(file_path)

# Convert date columns to datetime
data['Start date'] = pd.to_datetime(data['Start date'])
data['End date'] = pd.to_datetime(data['End date'])

# Extract latitude and longitude for start and end stations
stations = data[['Start station number', 'lat_start', 'lon_start']].drop_duplicates()
stations.columns = ['station_id', 'latitude', 'longitude']
```

```python
import seaborn as sns

# Count trips by station
trip_counts = data.groupby(['Start station number', 'lat_start', 'lon_start']).size().reset_index(name='trip_count')

# Plot heatmap of trip origins
plt.figure(figsize=(10, 8))
sns.scatterplot(x='lon_start', y='lat_start', size='trip_count', data=trip_counts, legend=False, sizes=(20, 200))
plt.title('Heatmap of Trip Origins')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Create a GeoDataFrame for stations
stations_gdf = gpd.GeoDataFrame(stations, geometry=gpd.points_from_xy(stations.longitude, stations.latitude))

# Plot current station locations
stations_gdf.plot(figsize=(10, 8), color='blue', markersize=5)
plt.title('Current Bike Sharing Station Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Example potential new locations (these would be derived from analysis)
new_stations = pd.DataFrame({
    'station_id': [4001, 4002, 4003],
    'latitude': [51.510, 51.515, 51.520],
    'longitude': [-0.120, -0.130, -0.135]
})
new_stations_gdf = gpd.GeoDataFrame(new_stations, geometry=gpd.points_from_xy(new_stations.longitude, new_stations.latitude))

# Plot current and potential new station locations
base = stations_gdf.plot(figsize=(10, 8), color='blue', markersize=5, label='Current Stations')
new_stations_gdf.plot(ax=base, color='red', markersize=5, label='New Stations')
plt.legend()
plt.title('Current and Potential New Bike Sharing Station Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.show()
```

```python
# Example function to calculate the average distance to the nearest station
from scipy.spatial import distance_matrix
import numpy as np

def average_distance_to_nearest_station(existing_stations, new_stations):
    all_stations = pd.concat([existing_stations, new_stations])
    dist_matrix = distance_matrix(all_stations[['latitude', 'longitude']], all_stations[['latitude', 'longitude']])
    np.fill_diagonal(dist_matrix, np.inf)
    nearest_distances = dist_matrix.min(axis=1)
    return nearest_distances.mean()

# Calculate average distance before and after adding new stations
before_avg_distance = average_distance_to_nearest_station(stations, pd.DataFrame())
after_avg_distance = average_distance_to_nearest_station(stations, new_stations)

print(f'Average distance to nearest station before: {before_avg_distance:.2f} meters')
print(f'Average distance to nearest station after: {after_avg_distance:.2f} meters')
```
