```python
import pandas as pd

# Load the data from the CSV file
file_path = 'bicycle_trip_data_3months.csv'  # replace with the actual path to your CSV file
df = pd.read_csv(file_path)
    
# Convert the 'Start date' column to datetime format
df['Start date'] = pd.to_datetime(df['Start date'])

# Filter out weekends (Saturday=5 and Sunday=6)
df = df[df['Start date'].dt.weekday < 5]

# Sort the data by 'Start date'
df = df.sort_values(by='Start date')

# Save the filtered and sorted data to a new CSV file
output_file_path = 'filtered_data.csv'  # replace with your desired output file path
df.to_csv(output_file_path, index=False)

print("Filtered and sorted data has been saved to", output_file_path)
```

```python
import pandas as pd

# Read the CSV file into a DataFrame
file_path = 'filtered_data.csv'  # Replace with the actual file path

try:
    df = pd.read_csv(file_path)
    # Display the first few rows of the DataFrame to understand its structure
    print(df.head())

    # Check if the DataFrame is empty
    if df.empty:
        print("The DataFrame is empty. Please check the CSV file for data.")
    else:
        # Use 'Total duration (ms)' for outlier detection
        duration_col = 'Total duration (ms)'

        # Remove outliers in the 'Total duration (ms)' column using the IQR method
        Q1 = df[duration_col].quantile(0.25)
        Q3 = df[duration_col].quantile(0.75)
        IQR = Q3 - Q1

        # Define lower and upper bounds for outliers
        lower_bound = Q1 - 1.5 * IQR
        upper_bound = Q3 + 1.5 * IQR

        # Filter the DataFrame to find outliers
        outliers = df[(df[duration_col] < lower_bound) | (df[duration_col] > upper_bound)]

        # Filter the DataFrame to remove outliers
        df_filtered = df[(df[duration_col] >= lower_bound) & (df[duration_col] <= upper_bound)]

        # Display the number of rows before and after filtering
        print(f"Number of rows before filtering: {len(df)}")
        print(f"Number of rows after filtering: {len(df_filtered)}")

        # Print out the entries that were removed
        if not outliers.empty:
            print("Entries removed as outliers:")
            print(outliers)
        else:
            print("No outliers were found and removed.")

        # Save the filtered DataFrame to a new CSV file
        output_file_path = 'cleaned_datafile.csv'  # Replace with the desired output file path
        df_filtered.to_csv(output_file_path, index=False)

        print("Outliers removed and filtered data saved to:", output_file_path)
except FileNotFoundError:
    print(f"The file at {file_path} was not found. Please check the file path and try again.")
except pd.errors.EmptyDataError:
    print("The file is empty. Please check the CSV file for data.")
except Exception as e:
    print(f"An error occurred: {e}")
```
