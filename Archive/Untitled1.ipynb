```python
import pandas as pd
import folium
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter

# Load the CSV file
file_path = 'Stations_Communities.csv'
data = pd.read_csv(file_path)

# Initialize geolocator
geolocator = Nominatim(user_agent="london_bike_sharing")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

# Geocode each station to get coordinates
data['Location'] = data['Station'].apply(geocode)
data['Latitude'] = data['Location'].apply(lambda loc: loc.latitude if loc else None)
data['Longitude'] = data['Location'].apply(lambda loc: loc.longitude if loc else None)

# Drop rows with missing coordinates
data.dropna(subset=['Latitude', 'Longitude'], inplace=True)

# Create a map centered around London
london_map = folium.Map(location=[51.5074, -0.1278], zoom_start=12)

# Define a color palette for different communities
colors = ['red', 'blue', 'green', 'purple', 'orange', 'darkred', 'lightred', 'beige', 
          'darkblue', 'darkgreen', 'cadetblue', 'darkpurple', 'white', 'pink', 'lightblue',
          'lightgreen', 'gray', 'black', 'lightgray']

# Add stations to the map
for _, row in data.iterrows():
    folium.CircleMarker(
        location=[row['Latitude'], row['Longitude']],
        radius=5,
        color=colors[row['Community'] % len(colors)],
        fill=True,
        fill_color=colors[row['Community'] % len(colors)],
        fill_opacity=0.7,
        popup=row['Station']
    ).add_to(london_map)

# Save map to an HTML file
london_map.save('london_bike_sharing_clusters.html')
```
